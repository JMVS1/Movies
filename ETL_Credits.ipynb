1).   Tomamos nuestro DataFrame existente llamado "df" y  se extraen los datos de la columna "crew"  Luego, se procesan los datos  y se crea un nuevo DataFrame llamado "crew_data" que contiene los datos desanidados del campo. Finalmente, se imprimen los primeros registros de "crew_data".

```python
import pandas as pd

if 'crew' in df.columns:
    crew_data = []
    for row in df['crew']:
        crew = eval(row)
        for member in crew:
            member['movie_id'] = df['id']
            crew_data.append(member)
    
    crew_data = pd.DataFrame.from_records(crew_data)
    
    # Imprimir los primeros registros del DataFrame desanidado
    crew_data.head(3)
```

Ahora vamos a desadinar el campo cast el cual se convierte en otro dataframe llamado cast_data

7).Tomamos nuestro DataFrame existente llamado "df" y  se extraen los datos de la columna "cast"  Luego, se procesan los datos  y se crea un nuevo DataFrame llamado "cast_data" que contiene los datos desanidados del campo. Finalmente, se imprimen los primeros registros de "cast_data".

```python
import ast
import pandas as pd

if 'cast' in df.columns:
    # Desanidar la columna 'cast'
    cast_data = []
    for row in df['cast']:
        cast_list = ast.literal_eval(row)
        for member in cast_list:
            member['movie_id'] = df['id']
            cast_data.append(member)

    cast_data = pd.DataFrame.from_records(cast_data)

    # Imprimir los primeros registros del DataFrame 'cast_data'
    print(cast_data.head())
```

8).Hacemos union del dataframe df y cast_data por que la informacion que contiene los dos dataframe es relevante para la consulta de actor que nos piden en el proyecto.

```python
import pandas as pd

# Unir los dataframes df y cast_data por la columna 'id'
merged1_df = df.merge(cast_data, on='id', how='inner')

# Verificar el resultado
merged1_df.head(3)
```
